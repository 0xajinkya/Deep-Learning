Importing The Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection And Processing

```python
# Loading The Data From Dataset
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
data_frame.describe()
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

1 -> Benign
0 -> Malignant

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
X.head()
```

```python
y.head()
```

Splitting The Data In Training And Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

**Building The Neural Network**

```python
# Importing Tensorflow And Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting Up The Layers Of NN

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')

])
```

```python
# Compiling The Neural Network

model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy']
                )
```

```python
# Training The Neural Network 

history = model.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

Standardized The Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.fit_transform(X_test)
```

Building The Model Again With Standardized Data

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Training The Neural Network
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy And Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.ylabel('Epoch')
plt.legend(['Training Data' , 'Validation Data'], loc='lower right')
```

Loss

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.ylabel('Epoch')
plt.legend(['Training Data' , 'Validation Data'], loc='lower right')
```

Accuracy Value The Model On Test Data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
y_pred = model.predict(X_test_std)
print(y_pred.shape)
print(y_pred)
```

First entry represents the probability of that the particular datapoint to be malignant and the second entry represents benign.
